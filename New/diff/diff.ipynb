```python
def find_diff(file1, file2):
    """
    Finds the lines present in file1 but not in file2, and vice versa.

    Args:
        file1 (str): Path to the first file.
        file2 (str): Path to the second file.

    Returns:
        tuple: A tuple containing two lists:
            - lines_only_in_file1: Lines present only in file1.
            - lines_only_in_file2: Lines present only in file2.
    """

    with open(file1, 'r') as f1:
        lines_file1 = set(line.strip() for line in f1)

    with open(file2, 'r') as f2:
        lines_file2 = set(line.strip() for line in f2)

    lines_only_in_file1 = list(lines_file1 - lines_file2)
    lines_only_in_file2 = list(lines_file2 - lines_file1)

    return lines_only_in_file1, lines_only_in_file2

file1 = 'downloaded_log_first.txt'
file2 = 'downloaded_log.txt'

diff1, diff2 = find_diff(file1, file2)

print(f"Lines only in {file1}:")
print(len(diff1))
for line in diff1:
    print(line)

print(f"\nLines only in {file2}:")

print(len(diff2))
for line in diff2:
    print(line)
```
