```python
import sys
import torch
import importlib.util

def check_library_compatibility():
    """
    Comprehensive diagnostic for library compatibility
    """
    print("🔍 Library Compatibility Diagnostic")

    # Python and PyTorch Information
    print("\n📋 System Information:")
    print(f"Python Version: {sys.version}")
    print(f"PyTorch Version: {torch.__version__}")
    print(f"CUDA Available: {torch.cuda.is_available()}")
    print(f"CUDA Version: {torch.version.cuda if torch.cuda.is_available() else 'Not Available'}")

    # Check MMCV Importability
    def check_library_import(library_name):
        print(f"\n🕵️ Checking {library_name} Import:")
        try:
            spec = importlib.util.find_spec(library_name)
            if spec is not None:
                print(f"✅ {library_name} is installed")
                module = importlib.import_module(library_name)
                print(f"   Version: {module.__version__ if hasattr(module, '__version__') else 'Unknown'}")
                print(f"   Path: {spec.origin}")
            else:
                print(f"❌ {library_name} not found")
        except Exception as e:
            print(f"❌ Error importing {library_name}: {e}")

    # Libraries to check
    libraries_to_check = [
        'mmcv',
        'mmdet',
        'torch',
        'torchvision'
    ]

    for lib in libraries_to_check:
        check_library_import(lib)

    # CUDA Extension Check
    def check_cuda_extensions():
        print("\n🧩 CUDA Extension Diagnostics:")
        try:
            import mmcv
            from mmcv.ops import RoIPool
            print("✅ MMCV CUDA Operations: Loaded Successfully")
        except ImportError as e:
            print(f"❌ CUDA Extension Import Failed: {e}")

    check_cuda_extensions()

def recommend_resolution():
    """
    Provide targeted resolution recommendations
    """
    print("\n🔧 Recommended Resolution Steps:")
    recommendations = [
        "1. Uninstall existing MMCV and MMDetection packages",
        "2. Install compatible versions matching your PyTorch:",
        "   pip install mmcv==2.0.0rc4 mmdet==3.0.0",
        "3. Ensure PyTorch, torchvision match CUDA version",
        "4. Consider using a virtual environment for isolation"
    ]

    for step in recommendations:
        print(step)

def main():
    check_library_compatibility()
    recommend_resolution()

if __name__ == "__main__":
    main()
```
